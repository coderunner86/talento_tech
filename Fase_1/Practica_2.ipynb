```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
#Creacion de datos artificiales
X, Y = make_circles(n_samples=500, factor=0.5, noise=0.05)
res = 100
#Declaracion de coordenadas e inputs
_x0 = np.linspace(-1.5, 1.5, res)
_x1 = np.linspace(-1.5, 1.5, res)
_pX = np.array(np.meshgrid(_x0, _x1)).T.reshape(-1,2)
_pY = np.zeros((res, res))+0.5
#Visualizacion
plt.figure(figsize=(8,8))
plt.pcolormesh(_x0, _x1, _pY, cmap="coolwarm", vmin=0, vmax=1)
plt.scatter(X[Y == 0,0], X[Y == 0,1], c="skyblue")
plt.scatter(X[Y == 1,0], X[Y == 1,1], c="salmon")
plt.tick_params(labelbottom = False, labelleft=False)
```

# **VISUALIZACION CON ScikitLearn**

```python
import sklearn as sk
import sklearn.neural_network as sknn
from IPython.core.display import display, HTML

lr = 'adaptive' #LEARNING RATE
nn = [2, 16, 8, 1] #NEURAL NUMBER BY LAYER

# Create the neural network multiplayer model object
clf = sk.neural_network.MLPRegressor(solver='sgd',
                          learning_rate=lr,
                          hidden_layer_sizes=tuple(nn[1:]),
                          verbose=True,
                          n_iter_no_change=1000,
                          batch_size=64)

# Train the model
clf.fit(X, Y)

```
